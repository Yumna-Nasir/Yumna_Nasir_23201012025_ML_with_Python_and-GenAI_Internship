```python
from google.colab import files
uploaded = files.upload()

#14. Linear Regression Model
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv("dataset.csv")
target_col = "yield_ton_per_hectare"
X = df.drop(columns=[target_col])
y = df[target_col]
X = X.select_dtypes(include=[np.number])
model = LinearRegression()
model.fit(X, y)
print(f"Model Intercept: {model.intercept_:.4f}\n")
coefficients_df = pd.DataFrame(
    {"Feature": X.columns, "Coefficient": model.coef_}
)
print("Model Coefficients:")
print(coefficients_df.to_string(index=False))
highest_positive_feature = coefficients_df.loc[coefficients_df["Coefficient"].idxmax()]
print(
    f"The feature with the highest positive coefficient is: "
    f"'{highest_positive_feature['Feature']}' "
    f"({highest_positive_feature['Coefficient']:.4f})"
)
```
